## Checkpoint 4: Sector Analysis 1

Extracting the main sector using the column ```category_list```. The category_list column contains values such as 'Biotechnology|Health Care' - in this, 'Biotechnology' is the 'main category' of the company, which we need to use.

Extracting the main categories in a new column.

````python
# extracting the main category
master_frame.loc[:, 'main_category'] = master_frame['category_list'].apply(lambda x: x.split("|")[0])
master_frame.head()
````

````python
# drop the category_list column
master_frame = master_frame.drop('category_list', axis=1)
master_frame.head()
````

````python
# read mapping file
mapping = pd.read_csv("mapping.csv", sep=",")
mapping.head()
````

````python
# missing values in mapping file
mapping.isnull().sum()
````

````python
# remove the row with missing values
mapping = mapping[~pd.isnull(mapping['category_list'])]
mapping.isnull().sum()
````

````python
# converting common columns to lowercase
mapping['category_list'] = mapping['category_list'].str.lower()
master_frame['main_category'] = master_frame['main_category'].str.lower()
````

````python
# look at heads
mapping.head()
````

````python
master_frame.head()
````

````python
mapping['category_list']
````

To merge all the ```main_category``` values with the mapping file's ```category_list``` column, values in the  ```main_category``` column should be present in the ```category_list``` column of the mapping file.

````python
# values in main_category column in df which are not in the category_list column in mapping file
master_frame[~master_frame['main_category'].isin(mapping['category_list'])]
````

Values such as 'analytics', 'business analytics', 'finance', 'nanatechnology' etc. are not present in the mapping file.

just checking the values which are present in the mapping file but not in the main dataframe ```master_frame```.

````python
# values in the category_list column which are not in main_category column 
mapping[~mapping['category_list'].isin(master_frame['main_category'])]
````

There are sectors named *alter0tive medicine*, *a0lytics*, *waste ma0gement*, *veteri0ry*, etc. 

Treating this problem by replacing '0' with 'na' in the ```category_list``` column.

````python
# replacing '0' with 'na'
mapping['category_list'] = mapping['category_list'].apply(lambda x: x.replace('0', 'na'))
print(mapping['category_list'])
````

````python
# merge the dfs
df = pd.merge(master_frame, mapping, how='inner', left_on='main_category', right_on='category_list')
df.head()
````

````python
# let's drop the category_list column since it is the same as main_category
df = df.drop('category_list', axis=1)
df.head()
````

## Checkpoint 5: Sector Analysis 2


### storing the value and id variables in two separate arrays

````python
# storing the value variables in one Series
value_vars = df.columns[9:18]

# taking the setdiff() to get the rest of the variables
id_vars = np.setdiff1d(df.columns, value_vars)

print(value_vars, "\n")
print(id_vars)
````

````python
# Sector wise  
melted_df = pd.melt(df, id_vars=list(id_vars), value_vars=list(value_vars))

melted_df.head()
````

````python
# remove rows having value=0
melted_df = melted_df[melted_df['value']==1]
melted_df = melted_df.drop('value', axis=1)
````

````python
# look at the new df
melted_df.head()
len(melted_df)
````

````python
# renaming the 'variable' column
melted_df = melted_df.rename(columns={'variable': 'sector'})
````

````python
# info
melted_df.info()
````

````python
# summarising the sector-wise number and sum of venture investments across three countries

# first, let's also filter for investment range between 5 and 15m
df = melted_df[(melted_df['raised_amount_usd'] >= 5000000) & (melted_df['raised_amount_usd'] <= 15000000)]
````

````python
# groupby country, sector and compute the count and sum
df.groupby(['country_code', 'sector']).raised_amount_usd.agg([ 'sum','count']).apply(lambda x: x.sort_values(ascending=False))
````

````python
# plotting sector-wise count and sum of investments in the three countries
plt.figure(figsize=(16, 14))

plt.subplot(2, 1, 1)
p = sns.barplot(x='sector', y='raised_amount_usd', hue='country_code', data=df, estimator=np.sum)
p.set_xticklabels(p.get_xticklabels(),rotation=30)
plt.title('Total Invested Amount (USD)')

plt.subplot(2, 1, 2)
q = sns.countplot(x='sector', hue='country_code', data=df)
q.set_xticklabels(q.get_xticklabels(),rotation=30)
plt.title('Number of Investments')


plt.show()
````

Thus, the top country in terms of the number of investments (and the total amount invested) is the USA. The sectors 'Others', 'Social, Finance, Analytics and Advertising' and 'Cleantech/Semiconductors' are the most heavily invested ones.

'News, Search and Messaging' is the next best sector in case we dont want to consider 'others'
